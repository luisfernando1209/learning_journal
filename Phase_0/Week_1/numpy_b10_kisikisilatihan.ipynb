Menampilkan angka dari list menggunakan loop dengan syarat:
- diberikan suatu list numbers = [12, 75, 150, 180, 145, 525, 50]
- Bilangan harus habis dibagi lima
- Jika jumlahnya lebih besar dari 150, lewati dan pindah ke nomor berikutnya
- Jika jumlahnya lebih besar dari 500, maka hentikan loop
-  keluaran yang diharapkan:
    ```
    75
    150
    145
    ```

````python
numbers = [12, 75, 150, 180, 145, 525, 50]

for i in numbers:
    if i > 500:
        break
    elif i > 150:
        continue
    elif i%5==0:
        print(i)
````

Hitunglah luas jika dual_sim = 1, jika tidak maka hitung kurang:

- luas = tinggi(px_height) * lebar(px_width)
- kurang = tinggi - lebar

````python
def luas(tinggi, lebar):
    hasil= tinggi * lebar
    return(hasil)

def kurang(tinggi, lebar):
    hasil_k= tinggi - lebar
    return(hasil_k)
````

````python
df_new = df.copy()
lk = []
j= 0

for i in df_new['dual_sim']:
    if i == 1:
        lk.append(luas(df_new['px_height'][j], df_new['px_width'][j]))
    else:
        lk.append(kurang(df_new['px_height'][j], df_new['px_width'][j]))
    j+=1
df_new['L/K'] = lk
````
