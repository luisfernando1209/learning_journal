### Column Manipulation
- Create new column
- Rename column
- Drop column

```python
df_new = df.copy()
df_new.head()
```

```python
# Create new columns
# Ada yang baru np.abs() buat ngeabsolutin angka biar minus juga jadi plus
df_new['selisih poin'] = np.abs(df_new['pts'] - df_new['opp_pts'])
df_new.head()
```

```python
# Rename columns
df_new.rename(columns={'game_location':'loc','game_result':'result'})
```

```python
df_new #cek apakah rename sudah tersimpan atau belum
```

```python
# Rename columns karena belum tersimpan , dikasi inplace
df_new.rename(columns={'game_location':'loc','game_result':'result'},inplace=True)
```

```python
#Cek hasil rename dengan inplace
df_new
```

```python
# Drop Column (ngedrop 1 kolom saja)
df_new.drop('elo_i',axis=1)
```

```python
# Drop Column (ngedrop 1 kolom saja)
df_new.drop(['elo_i','elo_n'],axis=1)
```

```python
# Drop Column (ngedrop 2 kolom ) -- kalau dah mau di hapus pake inplace
df_new.drop(['elo_i','elo_n'],axis=1,inplace=True)
```
